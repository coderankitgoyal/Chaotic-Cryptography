```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
```

```python
def keygen(x,r,size):
#here x define the initial conditions, r defines the control parameter and size defines the number of keys required
    key=[]

    for i in range(size):
        x=r*x*(1-x)
        key.append(int((x*pow(10,16))%256))
    return key
```

```python
# keygen(0.01,3.915,10)
```

```python
img = mpimg.imread('lion.jpg')
plt.imshow(img)
plt.show()
```

```python
# now we need to measure the heights of the image to know 
#how many number of keys are required 
h = img.shape[0]
w = img.shape[1]
key = keygen(0.01,3.95,h*w)
```

## Using Confusion Technique to Encrypting the image

```python
# substitution = xor ing pixel values with the key to get encrypted image
z=0
enimg =  np.zeros(shape=[h,w,3], dtype = np.uint8)
for i in range(h):
    for j in range(w):
        enimg[i,j] = img[i,j]^key[z]
        z+=1

plt.imshow(enimg)
plt.show()
```

### Here, we are symmetric encryption as the key for both encryption and decryption are the same. 

```python
#key = keygen(0.01,3.95,h*w)
z=0
decimg =  np.zeros(shape=[h,w,3], dtype = np.uint8)
for i in range(h):
    for j in range(w):
        decimg[i,j] = enimg[i,j]^key[z]
        z+=1

plt.imshow(decimg)
plt.show()
```
